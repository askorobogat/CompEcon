1. Find the roots of the solution of y  
2. Find the roots of the inverse solution  
3. Model and visualize three cases of stable system
4. Model and visualize three cases of unstable system

```python
# 1. (y = 0.2yt-1 + 0.35yt-2)
a1,a2 = 0.2,0.35
a,b,c = 1,-a1,-a2
d = (b)**2 - 4*(a*c)
r1 = (-b + np.sqrt(d))/(2*a)
r2 = (-b - np.sqrt(d))/(2*a)
print(r1)
print(r2)
```

```python
# 3. Case 1 stable

T = 100
a,b,c = 1,0.2,0.35
d = (-b)**2 - 4*(a*-c)
A = np.eye(T) + np.eye(T,k=-1)*-b + np.eye(T,k=-2)*-c
e = np.ones(T)
#e = np.random.normal(0,0.001, size=(T,1))
y = np.linalg.inv(A)@e
print(d)
plt.plot(y)
```

VAR(1):

$$y_{t} = a_{21} x_{t-1} + a_{22} y_{t-1}$$  

1. Find the roots of the solution of y  
2. Find the roots of the inverse solution  
3. Model and visualize pp. 289-290

```python
# 1.
a11,a22,a21,a12 = 0.7,0.7,-0.2,-0.2
a = np.array([[a11,a21],[a12,a22]])
print(a)
np.linalg.eig(a)
```

```python
# 2.
a,b,c = (a11*a22 - a12*a21),-(a11+a22),1
d = (-b)**2 - 4*(a*c)
r1 = (-b + np.sqrt(d))/(2*a)
r2 = (-b - np.sqrt(d))/(2*a)
print(r1)
print(r2)
```

```python
# 3.
e1 = np.random.normal(0,0.1, size=(T,1))
e2 = np.random.normal(0,0.1, size=(T,1))
T = 100
y = []
z = []
y.append(1)
z.append(1)
```

```python
a11,a22,a21,a12 = 0.7,0.7,0.2,0.2 # Stationary process 1: positive correlation
for t in range(1,T):
    a = a11*y[t-1] + a12*z[t-1] + e1[t,0]
    b = a21*y[t-1] + a22*z[t-1] + e2[t,0]
    y.append(a)
    z.append(b)
plt.plot(y, label = "line 1")
plt.plot(z, label = "line 2")
```

```python
a11,a22,a21,a12 = 0.5,0.5,-0.2,-0.2 # Stationary process 1: negative correlation
y = []
z = []
y.append(1)
z.append(1)
for t in range(1,T):
    a = a11*y[t-1] + a12*z[t-1] + e1[t,0]
    b = a21*y[t-1] + a22*z[t-1] + e2[t,0]
    y.append(a)
    z.append(b)
plt.plot(y, label = "line 1")
plt.plot(z, label = "line 2")
```

```python
a11,a22,a21,a12 = 0.5,0.5,0.5,0.5 # Unit root
y = []
z = []
y.append(1)
z.append(1)
for t in range(1,T):
    a = a11*y[t-1] + a12*z[t-1] + e1[t,0]
    b = a21*y[t-1] + a22*z[t-1] + e2[t,0]
    y.append(a)
    z.append(b)
plt.plot(y, label = "line 1")
plt.plot(z, label = "line 2")
```

```python
a11,a22,a21,a12 = 0.5,0.5,0.5,0.5 # Unit root + drift
y = []
z = []
y.append(1)
z.append(1)
for t in range(1,T):
    a = 0.5 + a11*y[t-1] + a12*z[t-1] + e1[t,0]
    b = a21*y[t-1] + a22*z[t-1] + e2[t,0]
    y.append(a)
    z.append(b)
plt.plot(y, label = "line 1")
plt.plot(z, label = "line 2")
```

```python
a10,a20,a11,a22,a21,a12 = 0,0.3,0.5,0.5,0.5,0.2 # SVAR
T = 1000
e1 = np.random.normal(0,0.1, size=(T,1))
e2 = np.random.normal(0,0.1, size=(T,1))
y = []
z = []
y.append(1)
y.append(1)
z.append(1)
z.append(1)
for t in range(1,T):
    yt = a10*z[t] + a11*y[t-1] + a12*z[t-1] + e1[t,0]
    zt = a20*y[t] + a21*y[t-1] + a22*z[t-1] + e2[t,0]
    y.append(yt)
    z.append(zt)
plt.plot(y, label = "line 1")
plt.plot(z, label = "line 2")
```

```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.tsa.api import VAR, SVAR

ar = np.array([y,z]).T
df = pd.DataFrame(ar,columns=['y','z'])
df
```

```python
# 3. SVAR Estimates

# Matrices A B
A = np.asarray([[1,0],['E',1]])
B = np.asarray([['E',0],[0,'E']])

# Model and results
mymodel = SVAR(df, svar_type='AB', A=A, B=B, missing='drop')
res = mymodel.fit(maxiter=10000, maxfun=10000)
```

```python
res.k_exog_user = 0
res.summary()
```

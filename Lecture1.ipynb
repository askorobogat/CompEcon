# Computational economics (CE)
## What is CE?

Use of computer to analyze (solve + simulate) complicated economic models that do not allow for analytical solutions

## Examples of CE

Include economics/business-related tasks which involve a lot of coding for such as

1. Computer algebra,
2. Data manipulation,
3. Data analysis using statistical methods (regression, machine learning, etc.)

## Why study CE?

The models are becoming more realistic and thus complicated:

Macroeconomics: heterogeneous agents, overlapping generations  
Microeconomics: dynamic games  
Structural econometrics: estimating complex models with micro-data  
Advanced econometrics: computationally intensive methods  

## Languages

Stat software: Stata, SPSS, R, Python, Matlab  
Math software: Mathematica, Maple

```python
# A function to calculate binomial coeffs: k from n
def binom(k, n):
    return factorial(n)//(factorial(k)*factorial(n-k))
```

## Linear systems of equations

### Market equilibrium in a linear model
When dealing with a number of markets and assuming linear demand and supply functions one can solve for market equilibrium

```python
# Let's define a function which creates a random matrix with properties needed for the problem

def random_matrix(n,positive=True,maxeigen=10):
    '''Generates square random positive/negative semi-definite matrix'''
    e=np.random.uniform(0,maxeigen,n) # random eigenvalues
    r=np.random.uniform(0,1,n*n).reshape(n,n) # rotation
    e = e if positive else -e
    A = np.diag(e)  # diagonal matrix with
    return r @ A @ r.T  # positive/negative semi-definite
```

```python
# Specify conditions

n = 3  # number of products
A = random_matrix(n,positive=False)  # demand
d = np.array([100,]*n)
B = random_matrix(n)  # supply
s = np.zeros(n)
print('Demand is given by Ap+d:\nA=%r\nd=%r' % (A,d))
print('Supply is given by Bp+s:\nB=%r\ns=%r' % (B,s))
```

```python
# Solution

p = np.linalg.solve(A-B, s-d)  # solve for equilibrium
q = A @ p + d # equilibrium quantities
print('Equilibrium prices are     p = {}'.format(p))
print('Equilibrium quantities are q = {}'.format(q))
```

```python
# 1-5 Egor Titov
import numpy as np
import pandas as pd
import re

data = pd.read_excel('https://github.com/askorobogat/CompEcon/raw/main/density.xls')

data = pd.melt(data,
                  id_vars=['Города'],
                  value_vars=[x for x in data.columns if x.isdigit()],
                  var_name='year',
                  value_name='popul')

data = data.rename(columns={'Города': 'city'})

data['popul'] = data['popul'].str.strip()
data['popul'] = data['popul'].apply(lambda x: np.nan if not x else x)

data = data.dropna()

data = data.sort_values(by='year')

data['popul'] = data['popul'].astype('float')

data.describe()

data.describe(include=['object'])

def percentile(n):
    def percentile_(x):
        return np.percentile(x, n)
    percentile_.name = 'percentile_%s' % n
    return percentile_

data.groupby(by='year').agg(['mean', 'std', 'min', 'max', percentile(25), percentile(50), percentile(75)])
```
